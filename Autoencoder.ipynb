```python
import keras
from keras import layers
import pandas as pd
```

```python
df_content = pd.read_csv("./datasets/cora/cora_content.csv")
```

```python
df_content
```

```python
df_content['label'].unique()
```

```python
df_content['label'] = df_content['label'].map({'Neural_Networks':0,'Rule_Learning': 1 ,'Reinforcement_Learning':2,'Probabilistic_Methods':3,'Theory':4,'Genetic_Algorithms':5,'Case_Based':6})
```

```python
df_content
```

```python
df_content.shape
```

```python
y = df_content['label']
df_content = df_content.drop(['label'],axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(df_content, y, train_size=0.75,random_state=40,stratify=y)

encoding_dim = 500

input_df = keras.Input(shape=(1434,))
encoded = layers.Dense(encoding_dim, activation='relu')(input_df)
decoded = layers.Dense(1434, activation='sigmoid')(encoded)

autoencoder = keras.Model(input_df, decoded)

encoder = keras.Model(input_df, encoded)

autoencoder.compile(optimizer='adadelta', loss='mean_squared_error')

autoencoder.fit(X_train, X_train,
                epochs=150,
                batch_size=256,
                shuffle=True,
                validation_data=(X_val, X_val))
```

```python
encoded_X_train = encoder.predict(X_train)
encoded_X_val = encoder.predict(X_val)
```
